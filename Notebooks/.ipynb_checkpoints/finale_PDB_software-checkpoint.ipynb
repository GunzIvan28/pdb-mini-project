#### Now Creating the read function ####

```python
global file
global special_list
special_list=[]
def read():
    import os.path 
    file=input("Please enter the name of the PDB file: ")
    path="/home/gunz/Ivan-Python/Notebooks/ivan-python-miniproject/Data/"
    full_path=path+file
    if os.path.exists(full_path) == True:
        print("Successfully loaded valid pdb file:",file)
        special_list.append(full_path)
#         print(special_list)
#         status from empty to 
    else:
        print("File or path '%s'" %(full_path), "does not exist \nFile should be located in the directory: %s" % path)
```

#### Now Creating the search function ####

```python
# hard coding for one chain pdb file# ---------> compilation working
def search():
    import re
#     global option
#     option = "s"
    if option.lower()=='s':
        count_list=[]    
        count=0  
        for i in special_list:      
            print("The loaded file(s) are:   \n")      
            print("  OPTION    FILE(s)")      
            print("    ", count ,"   ", i[62:])    
            count_list.append(str(count))  
        prompt=input("\nChoose one the files to work on from the options:")       

    OneCode = {'ALA':'A','CYS':'C','ASP':'D','GLU':'E','PCA':'E','PHE':'F','GLY':'G','HIS':'H','ILE':'I','LYS':'K','LEU':'L','MET':'M',
               'ASN':'N','PRO':'P','GLN':'Q','ARG':'R','SER':'S','THR':'T','VAL':'V','TRP':'W','TYR':'Y','LLP':'K'}




    blue = lambda text: '\033[0;34m' + text + '\033[0m' # Colors the printed output blue
    red = lambda text: '\033[0;31m' + text + '\033[0m'  # Colors the printed output red
    yellow = lambda text: '\033[0;33m' + text + '\033[0m'# Colors the printed output yellow
    green = lambda text: '\033[0;32m' + text + '\033[0m' # # Colors the printed output green

    with open(special_list[int(prompt)],"r") as insulin: 

        listA=[]

        for l in insulin:     
            if l.startswith("SEQRES"): #This only pulls out SEQRES and outputs it
                split_line = l.split()            
                if split_line[2] == "A":              
                    listA.append(split_line[4:])    


        flat_listA = []                           #create new list  
        for sublist in listA:                                                    
            for item in sublist:                  #2nd for loop considers each triplet code in each sublist and appends them to the list up        
                flat_listA.append(item)
    #     print(flat_listA)
        Aa_seqA="".join(OneCode[x] for x in flat_listA)
    #     print(Aa_seqA)
        print(red("ListA output is:"),Aa_seqA)    
        print(blue("######################################################################################################################################################\n"))

    pattern = re.compile(r'[GYL]A[PFW][TLVGMT]') #Find patterns starting with G or Y or L then A then P or F or W
    matches = pattern.finditer(Aa_seqA)
    count_matches = 0
    for match in matches:    
        count_matches+=1    
        print("Glycosylation site: ", count_matches, "---> position", match.span(),"---> match =", match.group() )
    #     print((str(match)))

    empty_aa = []
    lys_aa = list(Aa_seqA)


    pattern = re.compile(r'[GYL]A[PFW][TLVGMT]') #Find patterns starting with G or Y or L then A then P or F or W
    matches = pattern.finditer(Aa_seqA)
    c = 0
    for match in matches: 
        for i in range(int(match.start()), int(match.end())):        
            empty_aa.append(i)

    new_aa_lyst = []
    empty_aa
    #lys_aa
    c = 0
    for i in lys_aa:   
        if c in empty_aa:       
            new_aa_lyst.append(i.lower())   
        else:        
            new_aa_lyst.append(i)  
        c+=1
    print ("".join(new_aa_lyst))
```
